```python
# Getting the right info from the Scooby Doo Dataset
# scooby_use_df = all_data[['run.time', 'unmask.fred', 'unmask.daphne', 'unmask.velma', 'unmask.shaggy', 'unmask.scooby', 'my.glasses', 'jeepers', 'zoinks', 'jinkies', 'imdb']]
# series = all_data['series.name']
# # print(scooby_use_df)

# # scooby_use_df = all_data[['run.time', 'unmask.fred', 'unmask.daphne', 'unmask.velma', 'unmask.shaggy', 'unmask.scooby', 'my.glasses', 'jeepers', 'zoinks', 'jinkies']]
# # imdb = all_data['imdb']

# # Generate X & Y
# X = np.array(scooby_use_df)
# y = np.array(series)
# print(y)
test_data = all_data

all_data = all_data.dropna()

imdb_scores = all_data['imdb']

plt.figure(figsize=(10, 6))
plt.hist(imdb_scores, bins=150, range=(4, 10), edgecolor='k', alpha=0.7, color='green')
plt.title('IMDB Score Distribution of all_data after .dropna()')
plt.xlabel('IMDB Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()


# instead of dropping all rows with a single NAN(This significantly reduces the amount of data we can train with). 
    # Work on the columns individually to convert NAN to usable data

# check for nan in 'imdb' column
test_data.dropna(subset=['imdb'], inplace=True)

# Convert boolean strings to integers
bool_columns = ['caught.fred', 'caught.daphne', 'caught.velma', 'caught.shaggy', 'caught.scooby']

for column in bool_columns:
    test_data[column] = test_data[column].fillna(0)
    test_data[column] = test_data[column].replace({'TRUE': True, 'FALSE': False})
    test_data[column] = test_data[column].map({True: 1, False: 0, np.nan: np.nan}).astype('Int64')


# Convert "NULL" values to 0 
null_columns = ['my.glasses', 'jeepers', 'zoinks', 'jinkies']
for column in null_columns:
    test_data[column] = test_data[column].fillna(0)

# check columns to see if there are any NAN values
scooby_use_df_test = test_data[['run.time', 'caught.fred', 'caught.daphne', 'caught.velma', 'caught.shaggy', 'caught.scooby', 'my.glasses', 'jeepers', 'zoinks', 'jinkies', 'series.name']]

for column in scooby_use_df_test.columns:
    nan_values = scooby_use_df_test[column][scooby_use_df_test[column].isnull()]
    if not nan_values.empty:
        print("NaN found in column: ", {column})
        print(nan_values)


imdb_scores2 = test_data['imdb']

plt.figure(figsize=(10, 6))
plt.hist(imdb_scores2, bins=150, range=(4, 10), edgecolor='k', alpha=0.7, color='green')
plt.title('IMDB Score Distribution of test_data')
plt.xlabel('IMDB Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Uncomment to use test_data 
# all_data = test_data.copy()
```

### Test_data processing

```python
# Clean Up -- test_data
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder
from mpl_toolkits.mplot3d import Axes3D 
# Assuming 'series.name' is a categorical variable, we'll encode it
label_encoder = LabelEncoder()
scooby_use_df_test.loc[:, 'series.name'] = label_encoder.fit_transform(scooby_use_df_test['series.name'])
imdb_scores2 = test_data['imdb']

X2 = np.array(scooby_use_df_test) 
y2 = imdb_scores2

# Convert IMDb ratings into three classes: 'Low', 'Medium', and 'High'
bins = [0, 5, 7, np.inf]
labels = ['Low', 'Medium', 'High']
y_categories2 = pd.cut(y2, bins=bins, labels=labels)


# Ensure y_categories is of type 'str' or 'object'
y_categories2 = y_categories2.astype(str)

# test using test_data
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y_categories2, test_size=0.3, random_state=32)

print(y_train2)

class_counts = {'Low': 0, 'Medium': 0, 'High': 0}
for label in y_train2:
    class_counts[label] += 1

# Extract class labels and counts
labels = list(class_counts.keys())
counts = list(class_counts.values())
print(class_counts.values())


# Plotting the distribution
plt.figure(figsize=(6, 4))
plt.bar(labels, counts, color=['blue', 'orange', 'green'])
plt.xlabel('Classes')
plt.ylabel('Count')
plt.title('Class Distribution - Training Data')
plt.show()


print("Using testing data:")
for i in range(len(classifiers)):
    cn = names[i]
    clf = classifiers[i]
    clf.fit(X_train2, y_train2)
    y_predict = clf.predict(X_train2)
    train_acc = accuracy_score(y_train2, y_predict)
    y_predict = clf.predict(X_test2)
    test_acc = accuracy_score(y_test2, y_predict)
    print("Name:", cn)
    print("Training:", train_acc)
    print("Testing:", test_acc)
    print("\n")


pca = PCA(n_components=2)
pca.fit(X2)
X_new = pca.transform(X2)

# Create separate arrays for each class
X_low = X_new[y_categories2 == 'Low']
X_medium = X_new[y_categories2 == 'Medium']
X_high = X_new[y_categories2 == 'High']

# Scatter plot for each class separately
plt.scatter(X_low[:, 0], X_low[:, 1], label='Low', alpha=0.5)
plt.scatter(X_medium[:, 0], X_medium[:, 1], label='Medium', alpha=0.5)
plt.scatter(X_high[:, 0], X_high[:, 1], label='High', alpha=0.5)

plt.xlabel('Principal Component 1 (PCA-transformed)')
plt.ylabel('Principal Component 2 (PCA-transformed)')
plt.title('Scatter Plot of PCA Components with Class Distribution')
plt.legend()
plt.show()
```

### Linear SVC

```python
import warnings
from sklearn.svm import LinearSVC
from sklearn.exceptions import ConvergenceWarning

# Ignore FutureWarning
warnings.simplefilter(action='ignore', category=FutureWarning)

# Ignore ConvergenceWarning
warnings.simplefilter(action='ignore', category=ConvergenceWarning)

#Linear SVC - hyper-parameter tuning first 
max_iters = [1000, 10000, 100000, 1000000]
C_values = [0.01, 0.1, 1, 10, 100, 1000]


print("Trained with test_data")
#Create subtraining/validation set from test_data training set
X_subtrain2, X_val2, y_subtrain2, y_val2 = train_test_split(X_train2, y_train2, test_size=0.15, random_state=32)

validation_accuracies2 = []
param_combinations2 = []

i = 1
# Iterate through all combinations of max_iter and C
for iteration in max_iters:
    for C in C_values:
        clf = LinearSVC(max_iter=iteration, C=C, random_state=32)
        clf.fit(X_subtrain2, y_subtrain2)

        # Make predictions on the validation set
        y_pred2 = clf.predict(X_val2)

        # Calculate accuracy on the validation set
        accuracy2 = accuracy_score(y_val2, y_pred2)

        validation_accuracies2.append(accuracy2)
        param_combinations2.append((iteration, C))
        print(f"Combination:  {i},  Max Iterations= {iteration} ,  C= {C}  Accuracy={accuracy2}" )
        i = i+1
    

validation_accuracies_grid2 = np.array(validation_accuracies2).reshape(len(max_iters), len(C_values))

# Create a heatmap to visualize the results
plt.figure(figsize=(10, 6))
plt.imshow(validation_accuracies_grid2, cmap='viridis', origin='lower')
plt.colorbar()
plt.xticks(np.arange(len(C_values)), C_values)
plt.yticks(np.arange(len(max_iters)), max_iters)
plt.xlabel('C Values')
plt.ylabel('Max Iterations')
plt.title('Validation Accuracy  -- test_data')
plt.show()

print("Hyperparameters: max_iter=1000, C=1000 , random_state=32 ") 

clf = LinearSVC(max_iter=1000, C=1000, random_state=32)
clf.fit(X_train2, y_train2)

# Make predictions on the training set
train_pred = clf.predict(X_train2)
train_accuracy = accuracy_score(y_train2, train_pred)

print(f"Training Accuracy: {train_accuracy}")

# Make predictions on the testing set
test_pred = clf.predict(X_test2)
test_accuracy = accuracy_score(y_test2, test_pred)

print(f"Testing Accuracy:  {test_accuracy}")


print("Hyperparameters: max_iter=1000000, C=.01 , random_state=32 ") 

clf = LinearSVC(max_iter=1000000, C=.01, random_state=32)
clf.fit(X_train2, y_train2)

# Make predictions on the training set
train_pred = clf.predict(X_train2)
train_accuracy = accuracy_score(y_train2, train_pred)

print(f"Training Accuracy: {train_accuracy}")

# Make predictions on the testing set
test_pred = clf.predict(X_test2)
test_accuracy = accuracy_score(y_test2, test_pred)

print(f"Testing Accuracy:  {test_accuracy}")
```
